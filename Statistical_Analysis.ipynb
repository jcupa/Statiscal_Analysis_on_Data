## Computing the mean, trimmed_mean, median

### Price column  

```python
print(f"original mean: {data.price.mean()}")
print(f"median: {data.price.median()}")
print(f"trimmed mean: {stats.trim_mean(data.price,0.3)}")
```

The previous values confirm that there are not wide extremes at the end and begining of the range of the prices values. 

### Year column

```python
print(f"original mean: {data.year.mean()}")
print(f"median: {data.year.median()}")
print(f"trimmed mean: {stats.trim_mean(data.year,0.3)}")
```

The previous values confirm that there are not wide extremes at the end and begining of the range of the year values. 

### Mileage

```python
print(f"original mean: {data.mileage.mean()}")
print(f"median: {data.mileage.median()}")
print(f"trimmed mean: {stats.trim_mean(data.mileage,0.3)}")
```

The previous values confirm that there are wide extremes at the end and begining of the range of the mileage values. 

## Variability of data

### Standard deviation and Variance

#### Price column

```python
# Price column
print(f"std: {data.price.std()}")
print(f"variance: {data.price.var()}")
```

#### Year column

```python
print(f"std: {data.year.std()}")
print(f"variance: {data.year.var()}")
```

#### Mileage column

```python
print(f"std: {data.mileage.std()}")
print(f"variance: {data.mileage.var()}")
```

Both "Price" and "Mileage" columns have a very disperse data while "Year" columns does not vary to much.
